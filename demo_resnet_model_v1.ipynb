```python
import matplotlib.pyplot as plt 
import numpy as np 
import os 
import cv2
import PIL 
import pathlib 
import tensorflow as tf 
import matplotlib.pyplot as plt 
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.python.keras.layers import Dense, Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
img_width, img_height=497, 368
```

```python
model = keras.models.load_model("resnet_model_v1")
```

```python
img = cv2.imread("D:\\Lane_FYP\\Data_V1\\safe_driving\image-44.png")
plt.imshow(tf.image.resize(img, (256,256)).numpy().astype(int))
plt.show()
resize = cv2.resize(img, (img_height, img_width))
image = np.expand_dims(resize,axis=0)
pred=model.predict(image)
if pred>0.5:
    print("Reckless Driving")
else:
    print("Safe Driving")
```

```python
img2 = cv2.imread("D:\\Lane_FYP\\Data_V1\\unsafe_driving\\reckless_driving33.png")
plt.imshow(tf.image.resize(img2, (256,256)).numpy().astype(int))
plt.show()
resize2 = cv2.resize(img2, (img_height, img_width))
image2 = np.expand_dims(resize2,axis=0)
pred2=model.predict(image2)
if pred2>0.5:
    print("Reckless Driving")
else:
    print("Safe Driving")
```
